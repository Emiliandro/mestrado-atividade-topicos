### Média de preço por trimestre de 2020-2022

```python
import numpy as np
```

```python
def get_trimesters_median(data):
  filterd = data[['Trimester','District','Price','Year']].copy()
  cache2020 = {}
  cache2021 = {}
  cache2022 = {}

  for i in range(filterd['Trimester'].size):
    if (filterd['Year'][i] == 2020):
      try:
        cache2020[filterd['Trimester'][i]].append(filterd['Price'][i])
      except:
        cache2020[filterd['Trimester'][i]]= []
        cache2020[filterd['Trimester'][i]].append(filterd['Price'][i])

    if (filterd['Year'][i] == 2021):
      try:
        cache2021[filterd['Trimester'][i]].append(filterd['Price'][i])
      except:
        cache2021[filterd['Trimester'][i]]= []
        cache2021[filterd['Trimester'][i]].append(filterd['Price'][i])
    if (filterd['Year'][i] == 2022):
      try:
        cache2022[filterd['Trimester'][i]].append(filterd['Price'][i])
      except:
        cache2022[filterd['Trimester'][i]]= []
        cache2022[filterd['Trimester'][i]].append(filterd['Price'][i])
  mdfirst_trimester = []
  mdfirst_trimester.append(np.median(cache2020[1]))
  mdfirst_trimester.append(np.median(cache2021[1]))
  mdfirst_trimester.append(np.median(cache2022[1]))

  mdsecond_trimester = []
  mdsecond_trimester.append(np.median(cache2020[2]))
  mdsecond_trimester.append(np.median(cache2021[2]))
  mdsecond_trimester.append(np.median(cache2022[2]))
  return {'first':mdfirst_trimester,'second':mdsecond_trimester}
```

```python
trimesters = get_trimesters_median(df)
```

```python
x = np.arange(len(trimesters['second']))  # the label locations
labels = ['2020', '2021', '2022']
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, trimesters['first'], width, label='First')
rects2 = ax.bar(x + width/2, trimesters['second'], width, label='Second')
plt.gcf().set_size_inches(13, 7)

plt.xlabel("Trimestre")
plt.ylabel("Média de preços total")
ax.legend()

fig.tight_layout()
plt.show()
```
